```python
# Definition for number of points and grid creation
shape = (100, 100)
x, y, z = grids.regular_grid((-2000., 2000., -2000., 2000.), shape, -50.)
```

```python
# Plotting the result for approximated total field anomaly
plt.close('all')
plt.figure(figsize=(8,6))
plt.contourf(x.reshape(shape), y.reshape(shape), gz.reshape(shape), 20, cmap = plt.cm.RdBu_r)
plt.title('Gravity attraction (mGal)', fontsize = 14)
plt.xlabel('X coordinate (m)', fontsize = 14)
plt.ylabel('Y coordinate (m)', fontsize = 14)
plt.xticks(np.linspace(x.min(), x.max(), 6), fontsize = 12)
plt.yticks(np.linspace(y.min(), y.max(), 6), fontsize = 12)
plt.colorbar()
plt.show()
```
