```python
# importando as bibliotecas necessarias para o uso do script:
# import necessary library:
import numpy as np
import pylab as py
import pandas as pd
from mpl_toolkits.basemap import Basemap
# to plot figures in cells:
%matplotlib inline
```

```python
# bibliotecas particulares:
from codes.maps import map_plot
```

```python
# calculo da area onde serao plotadas as imagens:
long_min = -73.0 #np.min(lon_sat)
long_max = -30.0 #np.max(lon_sat)
lat_min = -50.0  #-33.0 #np.min(lat_sat)
lat_max = 10.0  #5.0 #np.max(lat_sat)

long_central = 0.5*(long_max + long_min)
lat_central = 0.5*(lat_max + lat_min)
area = [long_min, long_max, lat_min, lat_max]

escala = [-48.0, -44.0, -12.0, 5.0]
long0 = 0.5*(area[0] + area[1])
lat0 = 0.5*(area[2] + area[3])
print(long0,lat0)
```

```python
# projecao cartografica para o mapa:
mapa = Basemap(llcrnrlon=area[0],llcrnrlat=area[2],
            urcrnrlon=area[1],urcrnrlat=area[3],
            resolution='i',projection='poly',
            lon_0=long0,lat_0=lat0)
    
```

```python
# criacao do retangulo para destaque:
x0, y0 = mapa(-20.0,-10.0)
x1, y1 = mapa(0.0, 10.0)
rectangle = [x0,y0,x1,y1]
```

```python
# chamando funcao map_plot:
map_plot(mapa,area,"Studied Area",(18, 14), 10., estados=True, escala=rectangle, eixos=True)
```
