# Gravity attraction - Solid sphere

**Reference: Blakely (1996)**

```python
# Importing all usefull libraries
import numpy as np
import gravmag as gmc
import matplotlib.pyplot as plt
```

```python
# Positions and number of points
npts = 1000
x = np.linspace(-2000, 2000, npts)
y = np.copy(x)
z = -10.
# Computes the mesh of all points
X, Y = np.meshgrid(x, y)
```

```python
# Setting the parameters for a sphere
model1 = [-500., -750., 100., 200., 2500.] # meters + kg/m3
model2 = [800., 1200., 1250., 750., -2800.] # meters + kg/m3
model3 = [-650., 700., 450., 500., 2300.] # meters + kg/m3
```

```python
# Calculating the gravity attraction
gz = gmc.sphere_gz(X, Y, z, model1) + gmc.sphere_gz(X, Y, z, model2) + gmc.sphere_gz(X, Y, z, model3)
```

```python
from auxiliars import addnoise
```

```python
gz_noise = addnoise(gz, 0.5, 0.1)
```

```python
# Plotting the result for approximated total field anomaly
plt.close('all')
plt.figure(figsize=(16,6))

plt.subplot(1, 2, 1)
plt.contourf(X, Y, gz, 20, cmap = plt.cm.jet)
plt.title('Vertical gravity attraction (mGal)')
plt.xlabel('X coordinate (m)')
plt.ylabel('Y coordinate (m)')
plt.colorbar()

plt.subplot(1, 2, 2)
plt.contourf(X, Y, gz_noise, 20, cmap = plt.cm.jet)
plt.title('Corrupted gravity attraction (mGal)')
plt.xlabel('X coordinate (m)')
plt.ylabel('Y coordinate (m)')
plt.colorbar()

plt.show()
```
