# Gravitational potential and gravity attraction - Prism

**Reference: Blakely (1996)**

```python
# Importing all usefull libraries
import numpy as np
import gravmag as gmc
import matplotlib.pyplot as plt
```

```python
# Positions and number of points
npts = 500
x = np.linspace(-2000, 2000, npts)
y = np.copy(x)
z = np.random.normal(loc = -100., scale = 2., size = (npts))
# Computes the mesh of all points
X, Y = np.meshgrid(x, y)
```

```python
# Setting the parameters for prism
prism = [-500., 500., -500., 500., 250., 500., 2100] # meters + g/cm3
```

```python
# Calculating the gravitational potential
pot = gmc.potential(X, Y, z, prism)
```

```python
# Calculating the gravity components
gx = gmc.prism_gx(X, Y, z, prism)
gy = gmc.prism_gy(X, Y, z, prism)
gz = gmc.prism_gz(X, Y, z, prism)
```

```python
# Ploting all results 
plt.close('all')
plt.figure(figsize=(20,14))

plt.subplot(2, 2, 1)
plt.title('Gravitational potential (Joule/kg)')
plt.contourf(X, Y, pot, 20, cmap = plt.cm.jet)
plt.colorbar()
plt.xlabel('X coordinate (m)')
plt.ylabel('Y coordinate (m)')

plt.subplot(2, 2, 2)
plt.title('Gravity $X$ component (mGal)')
plt.contourf(X, Y, gx, 20, cmap = plt.cm.jet)
plt.colorbar()
plt.xlabel('X coordinate (m)')
plt.ylabel('Y coordinate (m)')

plt.subplot(2, 2, 3)
plt.title('Gravity$Y$ component (mGal)')
plt.contourf(X, Y, gy, 20, cmap = plt.cm.jet)
plt.colorbar()
plt.xlabel('X coordinate (m)')
plt.ylabel('Y coordinate (m)')

plt.subplot(2, 2, 4)
plt.title('Gravity $Z$ component (mGal)')
plt.contourf(X, Y, gz, 20, cmap = plt.cm.jet)
plt.colorbar()
plt.xlabel('X coordinate (m)')
plt.ylabel('Y coordinate (m)')

plt.show()
```
