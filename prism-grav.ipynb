# Gravitational potential and gravity attraction - Prism

**Reference: Blakely (1996)**

```python
# Importing all usefull libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Positions and number of points
n = 500
x = np.linspace(-2000, 2000, n)
y = np.copy(x)
z = -100.
# Computes the mesh of all points
X, Y = np.meshgrid(x, y)
```

```python
# Setting the parameters for prism
model = [-500., 500., -500., 500., 250., 500., 2100] # meters + g/cm3
```

```python
from codes.prism import potential, prism_gx, prism_gy, prism_gz
```

```python
# Calculating the gravitational potential
pot = potential(X, Y, z, model)
```

```python
# Calculating the gravity components
gx = prism_gx(X, Y, z, model)
gy = prism_gy(X, Y, z, model)
gz = prism_gz(X, Y, z, model)
```

```python
# Ploting all results 
plt.close('all')
plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
plt.title('Gravitational potential (Joule/kg)')
plt.contourf(X, Y, pot, 20, cmap = plt.cm.jet)
plt.xlabel('X (meters)', fontsize = 12)
plt.ylabel('Y (meters)', fontsize = 12)
plt.xticks(np.linspace(x.min(), x.max(), 5))
plt.yticks(np.linspace(y.min(), y.max(), 5))
plt.colorbar()

plt.subplot(2, 2, 2)
plt.title('Gravity $X$ component (mGal)')
plt.contourf(X, Y, gx, 20, cmap = plt.cm.jet)
plt.xlabel('X (meters)', fontsize = 12)
plt.ylabel('Y (meters)', fontsize = 12)
plt.xticks(np.linspace(x.min(), x.max(), 5))
plt.yticks(np.linspace(y.min(), y.max(), 5))
plt.colorbar()

plt.subplot(2, 2, 3)
plt.title('Gravity $Y$ component (mGal)')
plt.contourf(X, Y, gy, 20, cmap = plt.cm.jet)
plt.xlabel('X (meters)', fontsize = 12)
plt.ylabel('Y (meters)', fontsize = 12)
plt.xticks(np.linspace(x.min(), x.max(), 5))
plt.yticks(np.linspace(y.min(), y.max(), 5))
plt.colorbar()

plt.subplot(2, 2, 4)
plt.title('Gravity $Z$ component (mGal)')
plt.contourf(X, Y, gz, 20, cmap = plt.cm.jet)
plt.xlabel('X (meters)', fontsize = 12)
plt.ylabel('Y (meters)', fontsize = 12)
plt.xticks(np.linspace(x.min(), x.max(), 5))
plt.yticks(np.linspace(y.min(), y.max(), 5))
plt.colorbar()

plt.savefig('figures/prism-gravity', facecolor='w', edgecolor='w', bbox_inches='tight')
plt.show()
```
